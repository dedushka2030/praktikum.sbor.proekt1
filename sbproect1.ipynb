```python
data = data.dropna(subset=['name', 'year_of_release', 'genre'])
```

```python
data['rating'] = data['rating'].fillna('unknown')
```

*В столбце `user_score` имеется аббревиатура "tbd". Поиск данной аббревиатуры в интернете показал, что расшифровывается она как "to be decided" и переводится "не определено" или "будет определено позже", посмотрим какое количество строк с такой аббревиатурой*

```python
data.query('user_score == "tbd"')['user_score'].count()
```

*Всего 2376 строк имеют данную аббревиатуры в столбце `user_score`, это достаточно большое количество чтобы удалить их, но для дальнейших расчетов со столбцом `user_score` нам необходимо иметь числовой тип данных, поэтому заменим "tbd" и пропуски на значение `-1`.
```python
data['user_score'] = data['user_score'].replace('tbd', -1).fillna(-1)
```

```python
data['critic_score'] = data['critic_score'].fillna(-10)
```

#### Посмотрите, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Постройте диаграмму рассеяния и посчитайте корреляцию между отзывами и продажами. Сформулируйте выводы.

*Для построения диаграммы рассеяния возьмем платформу `PS4` и посмотрим как влияют на продажи отзывы пользователей и критиков, а так же посчитаем коэффициент корреляции между ними*

### Шаг 5. Проверьте гипотезы

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

*Нулевая гипотеза: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые. Альтернативная гипотеза: Средние пользовательские рейтинги платформ Xbox One и PC разные. Уровень статистической значимости выберем 0,01.*

```python
# создадим новы переменные х1 и х2, в которых сохраним пользывательский рейтинг по платформам XOne и PC
x1 = data_after_2012[(data_after_2012['platform'] == 'XOne') & (data_after_2012['user_score'] != -1)]['user_score']
x2 = data_after_2012[(data_after_2012['platform'] == 'PC') & (data_after_2012['user_score'] != -1)]['user_score']

alpha = 0.01 # уровень статистической значимости
# если p-value окажется меньше него, отвергнем гипотезу

# выберем метод для проверки гипотезы о равенстве среднего
results = st.ttest_ind(x1, x2)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

***Полученное значение p-value больше заданного уровня значимости, поэтому нулевую гипотезу мы не отвергаем. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.***

#### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные. Задайте самостоятельно пороговое значение alpha.

*Нулевая гипотеза: Средние пользовательские рейтинги жанров Action и Sports равны. Альтернативная гипотеза: Средние пользовательские рейтинги жанров Action и Sports разные. Уровень статистической значимости выберем 0,01.*

```python
# создадим новы переменные у1 и у2, в которых сохраним пользывательский рейтинг по жанрам Action и Sports
y1 = data_after_2012[(data_after_2012['genre'] == 'Action') & (data_after_2012['user_score'] != -1)]['user_score']
y2 = data_after_2012[(data_after_2012['genre'] == 'Sports') & (data_after_2012['user_score'] != -1)]['user_score']

alpha = 0.01 # уровень статистической значимости
# если p-value окажется меньше него, отвергнем гипотезу

# выберем метод для проверки гипотезы о равенстве среднего
results = st.ttest_ind(y1, y2)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

***Полученное значение p-value меньше заданного уровня значимости, поэтому нулевую гипотезу мы отвергаем. Есть основания утверждать, что средние пользовательские рейтинги жанров Action и Sports разные.***
